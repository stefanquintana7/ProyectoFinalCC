```python
import numpy as np
import random as rd


#Establecemos los valores de x
xar = [(-1 + i*0.001) for i in range(2000)]

#Establecemos los valores de y
yar = [e^-j^2 for j in xar]

#Repeticiones Montecarlo
n = 1000

#Hacer cuadrado que deje inscrito
x1 = -1
x2 = 1
y1 = min(yar)
y2 = max(yar)

#Método Montecarlo
i = 0
#Contar casos de éxito
exito = 0

#plot de puntos

while i < n:
    #Generar los puntos
    x0 = rd.uniform(x1,x2)
    y0 = rd.uniform(y1,y2)

    #Criterio de evaluación
    val = e^-x0^2
    if val > 0:
        if y0 < val and y0 > 0:
            exito += 1
    elif val < 0:
        if y0 > val and y0 < 0:
            exito += 1
            
    i += 1

#Calcular la integral
Integral = 2*(x2-x1)*(y2 - y1)*(exito/n)
print(Integral)

plot(e^-x^2, (-1,1), color = 'green')
```
